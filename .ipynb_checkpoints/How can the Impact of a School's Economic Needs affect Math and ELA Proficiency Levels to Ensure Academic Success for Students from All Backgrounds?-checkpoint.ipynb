**Data Cleaning**

```python
# Imported Relevant Modules
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)

# Retrieved Relevant Dataset and Created DataFrame
schooldata = pd.DataFrame(pd.read_csv("/Users/oliverkeyes/Dropbox/UofT Year 2/ECO225/Project/Data/2016 School Explorer.csv"))

# Edited DataFrame to Include Only Necessary Columns
schooldata = schooldata[["Economic Need Index", "School Income Estimate", "Percent Asian", "Percent White", "Percent Black / Hispanic", "Average ELA Proficiency", "Average Math Proficiency"]]

# Cleaned Percentile Columns and Converted to Integer Types
schooldata["Percent Asian"] = schooldata["Percent Asian"].str.replace("%", "").astype(int)
schooldata["Percent White"] = schooldata["Percent White"].str.replace("%", "").astype(int)
schooldata["Percent Black / Hispanic"] = schooldata["Percent Black / Hispanic"].str.replace("%", "").astype(int)

# Cleaned School Income Estimate Column and Converted to Float Type
schooldata["School Income Estimate"] = schooldata["School Income Estimate"].str.replace("$", "", regex=True)
schooldata["School Income Estimate"] = schooldata["School Income Estimate"].replace(",", "", regex=True)
schooldata["School Income Estimate"] = schooldata["School Income Estimate"].astype(float)

# Created a New Column Percent White / Asian to Contrast Percent Black / Hispanic Column
schooldata["Percent White / Asian"] = schooldata["Percent Asian"] + schooldata["Percent White"]

# Removed Unnecessary Columns from DataFranme
schooldata = schooldata.drop(columns=["Percent Asian", "Percent White"])
```

```python
# Removed Schools With Unavailable Data
schooldata = schooldata.dropna()

# Made DataFrame according to Economic Data
ecodata = schooldata[["Economic Need Index", "School Income Estimate", "Average ELA Proficiency", "Average Math Proficiency"]]

# Categorized Schools by Race Percentiles
schooldata["Black/Hispanic Dominant Schools"] = schooldata["Percent Black / Hispanic"] >= 70
schooldata["White/Asian Dominant Schools"] = schooldata["Percent White / Asian"] >= 70
schooldata["Mixed Race Schools"] = ((schooldata["Black/Hispanic Dominant Schools"]==False) & (schooldata["White/Asian Dominant Schools"]==False))

# Made DataFrames according to Race Percentiles
bhdata = schooldata[schooldata["Black/Hispanic Dominant Schools"]][["Average ELA Proficiency", "Average Math Proficiency", "Economic Need Index", "School Income Estimate"]]
wadata = schooldata[schooldata["White/Asian Dominant Schools"]][["Average ELA Proficiency", "Average Math Proficiency", "Economic Need Index", "School Income Estimate"]]
mrdata = schooldata[schooldata["Mixed Race Schools"]][["Average ELA Proficiency", "Average Math Proficiency", "Economic Need Index", "School Income Estimate"]]
```

**Justification for Assistance**

As previously said, education is a vital aspect of a child's success. One of the primary purposes of PASSNYC is to provide kids from all backgrounds with equal opportunity for educational excellence. To further the goal of analyzing the impact of schools' and students' economic needs on proficiency levels, a report is prepared for schools with varying race percentiles. Identifying Black/Hispanic dominant schools as those with a Black/Hispanic student population percentage greater than or equal to 70% and White/Asian dominant schools as those with a White/Asian student population percentage greater than or equal to 70%. With this information we are able to better assess the benefits and drawbacks of various groups. Furthermore, schools that were classified as neither were recognized as mixed-race schools to serve as a control.

```python
# Created a Density Plot Comparing Average ELA Proficiency by Race Percentiles
plot1 = bhdata["Average ELA Proficiency"].plot.density(figsize=(7,3))
plot1 = wadata["Average ELA Proficiency"].plot.density(figsize=(7,3))
plot1 = mrdata["Average ELA Proficiency"].plot.density(figsize=(7,3))
plot1 = plt.title("Comparing Average ELA Proficiency by Race Dominant Schools")
plot1 = plt.legend(["Black/Hispanic", "White/Asian", "Mixed Race"])
plot1 = plt.xlabel("Average ELA Proficiency")
plot1 = plt.ylabel("Frequency")
```

```python
# Created a Density Plot Comparing Average Math Proficiency by Race Percentiles
plot2 = bhdata["Average Math Proficiency"].plot.density(figsize=(7,3))
plot2 = wadata["Average Math Proficiency"].plot.density(figsize=(7,3))
plot2 = mrdata["Average Math Proficiency"].plot.density(figsize=(7,3))
plot2 = plt.title("Comparing Average Math Proficiency by Race Dominant Schools")
plot2 = plt.legend(["Black/Hispanic", "White/Asian", "Mixed Race"])
plot2 = plt.xlabel("Average Math Proficiency")
plot2 = plt.ylabel("Frequency")
```

**Analysis/Interpretation** 

Whenthe two density plots above, which compare average ELA and Math proficiency by race dominant schools, are analyzed, it is clear that White/Asian dominant schools have higher ELA and Math proficiency levels than Black/Hispanic dominant schools. Furthermore, an intriguing finding is that the Math and ELA performance levels of mixed-race schools appear to be in the midrange of the dominant race schools. As a result, the findings provide even more impetus to investigate the economic variables affecting each institution.

```python
# Created a Density Plot Comparing Economic Need Indexes by Race Percentiles
plot3 = bhdata["Economic Need Index"].plot.density(figsize=(7,3))
plot3 = wadata["Economic Need Index"].plot.density(figsize=(7,3))
plot3 = mrdata["Economic Need Index"].plot.density(figsize=(7,3))
plot3 = plt.title("Comparing Economic Need Indexes by Race Dominant Schools")
plot3 = plt.legend(["Black/Hispanic", "White/Asian", "Mixed Race"])
plot3 = plt.xlabel("Economic Need Indexes")
plot3 = plt.ylabel("Frequency")
```

```python
# Created a Density Plot Comparing School Income Estimates by Race Percentiles
plot4 = bhdata["School Income Estimate"].plot.density(figsize=(7,3))
plot4 = wadata["School Income Estimate"].plot.density(figsize=(7,3))
plot4 = mrdata["School Income Estimate"].plot.density(figsize=(7,3))
plot4 = plt.title("School Income by Race Dominant Schools")
plot4 = plt.legend(["Black/Hispanic", "White/Asian", "Mixed Race"])
plot4 = plt.xlabel("School Income Estimates")
plot4 = plt.ylabel("Frequency")
```

**Analysis/Interpretation** 

When the two density charts above are compared, it is evident that White/Asian dominant schools have lower economic need indices and greater school income estimates than Black/Hispanic dominant schools. Furthermore, a noteworthy discovery is that mixed-race schools' economic need indexes and school income estimations appear to be in the midpoint of the dominant race schools. As a result, addressing the economic needs of schools with high economic indexes and low incomes will help underrepresented groups to flourish academically and offer students from all backgrounds with equitable chance for educational excellence.
